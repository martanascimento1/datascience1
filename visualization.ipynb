```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 30])
y = np.array([0, 300])

plt.subplot(1, 2,1)
plt.plot(x, y)                

x2 = np.array([-3, -1, 8, 10])
y2 = np.array([7, 12, 0, 5])
plt.subplot(1, 2, 2)
plt.plot(x2, y2)
```
